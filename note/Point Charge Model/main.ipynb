```python
import math

# Permittivity of free space (ε₀) in F/m
eps_o = 8.854e-12  

# Radius in meters (distance from point charge)
radius = 5 #0.1  # m  

# Charge in Coulombs
charge = 1  # C  

# Print input values
print("Permittivity (ε₀):", f"{eps_o:.3e}")
print("Radius (r):", f"{radius:.3e} m")
print("Charge (Q):", f"{charge:.3e} C")

# Electric field due to a point charge:
# E = Q / (4 * π * ε₀ * r²)
electric_field = charge / (4 * math.pi * eps_o * radius**2)

# Print result in scientific notation
print("Electric Field (E):", f"{electric_field:.3e} N/C")
```
